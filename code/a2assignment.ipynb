```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hdf = pd.read_json('../data/historical data.json')
hdf = hdf.T
```

```python
hdf
```

```python
quiz_info = []
for i in hdf:
    quiz_info.append(hdf[i]['quiz'])
```

```python
import datetime

pd.to_datetime(hdf[0]['quiz']['end_time']) - pd.to_datetime(hdf[0]['quiz']['created_at'])
```

```python
quizdf = pd.DataFrame(quiz_info).T
hdf = pd.concat([quizdf, hdf])
```

```python
cols = ['description', 'quiz_id', 'user_id', 'name', 'difficulty_level', 'is_published', 'negative_marks', 'correct_answer_marks','shuffle',
       'show_answers', 'lock_solutions', 'is_form', 'show_mastery_option',
       'reading_material', 'quiz_type', 'is_custom', 'banner_id', 'exam_id',
       'show_unanswered', 'ends_at', 'lives', 'live_count', 'coin_count', 'daily_date',
       'reading_materials', 'id', 'quiz_id', 'user_id', 'submitted_at',
       'created_at', 'updated_at', 'quiz', 'response_map', 'time', 'end_time', 'better_than']
hdf = hdf.T
hdf.drop(columns=cols, inplace=True)
hdf = hdf.T
```

```python
hdf.head()
```

```python
ohdf = hdf.T
ohdf.head()
```

```python
time_took = []
for i in (ohdf['ended_at'] - ohdf['started_at']).tolist():
    time_took.append(int(i.total_seconds()/60))
ohdf['time_took'] = time_took
```

```python
ohdf['rank_text'] = ohdf['rank_text'].apply(lambda x: x.split('-')[-1].replace('#', ''))
```

```python
ohdf['accuracy'] = ohdf['accuracy'].apply(lambda x: x.split('%')[0])
```

```python
ohdf['total_questions'] = ohdf['incorrect_answers'] + ohdf['correct_answers'] 
ohdf['percentage'] = (ohdf['final_score'] / (ohdf['total_questions'] * 4) * 100)
```

```python
ohdf.drop(columns=["duration", "questions_count", "title", "type", "rank_text", "trophy_level", "score"], inplace=True)
```

```python
ohdf
```

```python
cols_to_convert = ohdf.columns.difference(["topic", "source", "started_at", "ended_at"])
ohdf[cols_to_convert] = ohdf[cols_to_convert].apply(pd.to_numeric)
ohdf['topic'] = ohdf['topic'].apply(lambda x : x.strip().lower())
```

```python
topicwise = ohdf.groupby('topic')
```

```python
topicwise_percent = topicwise.agg({'percentage': 'mean'})
plt.figure(figsize=(10, 8))
sns.barplot(x=topicwise_percent.index, y=topicwise_percent['percentage'], hue=topicwise_percent.index, legend=True)
plt.xticks([])
plt.ylabel("Percentage")
plt.title("Topic-wise Percent")
plt.tight_layout()
plt.show()
```

```python
topicwise_accuracy = topicwise.agg({'accuracy': 'mean'})
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x=topicwise_accuracy.index, y=topicwise_accuracy['accuracy'], hue=topicwise_accuracy.index, legend=True)
plt.xticks([])
plt.ylabel("Accuracy (%)")
plt.title("Topic-wise Accuracy")
plt.tight_layout()
plt.show()
```

```python
topicwise_time_took = topicwise.agg({'time_took': 'mean'})
plt.figure(figsize=(10, 8))
sns.barplot(x=topicwise_time_took.index, y=topicwise_time_took['time_took'], hue=topicwise_time_took.index, legend=True)
plt.xticks([])
plt.ylabel("Time (span of 15 minutes)")
plt.title("Topic-wise Time Took")
plt.tight_layout()
plt.show()
```

```python
topicwise_speed = topicwise.agg({'speed': 'mean'})
plt.figure(figsize=(10, 8))
sns.barplot(x=topicwise_speed.index, y=topicwise_speed['speed'], hue=topicwise_speed.index, legend=True)
plt.xticks([])
plt.ylabel("Speed")
plt.title("Topic-wise Speed")
plt.tight_layout()
plt.show()
```

```python
date = pd.to_datetime(ohdf['started_at'])
relative_day = []
for i, v in enumerate(date):
    relative_day.append((date[0].date() - date[i].date()).days)
ohdf['relative_day'] = relative_day
```

```python
ohdf
```

```python
topicwise_days_dfs = []
for col in topicwise.groups:
    topicwise_days_dfs.append(topicwise.get_group(col))
    
def plot_topicwise_relative_to_days_passed(column):
    num_topics = len(topicwise_days_dfs)
    cols = 2  # Number of columns in the subplot grid
    rows = (num_topics + cols - 1) // cols #Calculate the number of rows

    fig, axes = plt.subplots(rows, cols, figsize=(15, 8), squeeze=False)

    for i, ax in enumerate(axes.flat):
        if i < num_topics:
            sns.barplot(
                x=topicwise_days_dfs[i]['relative_day'],
                y=topicwise_days_dfs[i][column],
                hue=topicwise_days_dfs[i]['relative_day'],
                ax=ax,
                legend=False
            )
            ax.set_ylabel(column)
            ax.set_title(f"{column} Relative to Days Passed (Topic: {topicwise_days_dfs[i]['topic'].unique()[0]})")
            ax.set_xlabel("Relative Day")
        else:
            ax.axis("off")  # Hide extra subplots if any

    plt.tight_layout()
    plt.show()
```

```python
plot_topicwise_relative_to_days_passed('accuracy')
```

```python
plot_topicwise_relative_to_days_passed('time_took')
```
